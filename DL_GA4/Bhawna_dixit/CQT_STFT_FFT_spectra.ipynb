```python
import librosa
import librosa.display
import pandas as pd
import os 
import matplotlib.pyplot as plt 
import numpy as np
window = 30
path2 = "/content/musicnet/musicnet/"


meta_data=pd.read_csv('/content/musicnet_metadata.csv')
meta_data.head()

path = os.getcwd()
idx= []
for i in meta_data["id"]:
    idx.append(i)
# print(idx)
if not os.path.exists(path+"/Spectograms_cqt_train"):
    os.mkdir(path+ "/Spectograms_cqt_train")

for i in idx:
    print(i)
    try:
      if os.path.exists(path2 +"/train_data"+ os.sep+ str(i)+".wav"):
          file= path2 + "/train_data"+ os.sep+ str(i)+".wav"
          samples, sample_rate = librosa.load(file, sr=None)
          for j in range(len(samples)//(window*sample_rate)):
              X1 = librosa.cqt(samples[j*window*sample_rate:(j+1)*window*sample_rate])
              Xdb = librosa.amplitude_to_db(abs(X1), ref=np.max)
              plt.figure(figsize=(14, 5))
              librosa.display.specshow(Xdb, sr=sample_rate, x_axis='time', y_axis='hz')

              temp = meta_data[meta_data['id']==int(1727)]
              composer = list(temp['composer'])[0]
              if not os.path.exists(path +"/Spectograms_cqt_train/" + os.sep  + composer):
                  os.mkdir(path +"/Spectograms_cqt_train/" + os.sep + composer)
              if not os.path.exists(path +"/Spectograms_cqt_train/" + os.sep + composer + os.sep + str(i)):
                  os.mkdir(path +"/Spectograms_cqt_train/" + os.sep  + composer + os.sep + str(i))
              plt.savefig(path + "/Spectograms_cqt_train/" + os.sep + composer + os.sep + str(i) + os.sep + format(i)+'_'+str(j)+".png",transparent = True)
              plt.close()
    except:
      pass
```
